````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
````

**Graficos de lineas: plot**

La función plot() de Matplotlib se utiliza para generar gráficos de líneas  para visualizar datos.

Los gráficos de líneas permiten visualizar los cambios a lo largo de un rango continuo, como el tiempo o la distancia. La visualización del cambio con un gráfico de líneas permite ver de una sola vez la tendencia general y comparar simultáneamente varias tendencias.

**Grafica a partir de un array**



```
plt.plot(x, y, estilo_formato, ...)

```



````python
array= np.array([2,3,4,1,6,7,5,1])
plt.plot(array)
plt.xlabel('indice') #etiqueta eje x
plt.ylabel('Elemento') #etiqueta eje y
plt.title('Gráfico de ejemplo')  # Título del gráfico
plt.grid(True)        # Agregar una cuadrícula al gráfico
plt.show()

````

Por defecto, .plot(array) une con una línea la secuencia de puntos de coordenadas (indice,array[indice]) siendo ((0,2),(1,3) ......(7,1))
 los índices válidos de la lista desde 0


````python

ventas = {
    2010: 150000,
    2011: 180000,
    2012: 200000,
    2013: 230000,
    2014: 210000,
    2015: 250000,
    2016: 270000,
    2017: 300000,
    2018: 320000,
    2019: 280000,
    2020: 200000,
}

# Definimos la figura y los ejes del gráfico con Matplotlib
#Figura y conjuntos de ejes
fig, ax = plt.subplots()

# Dibujamos la línea con las ventas a partir del diccionario
ax.plot(ventas.keys(), ventas.values())

# Configuramos etiquetas para los ejes x e y
ax.set_xlabel('Año')
ax.set_ylabel('Ventas')

# Añadimos un título al gráfico
ax.set_title('Ventas anuales')

# Mostramos el gráfico por pantalla
plt.show()
````

La función subplots() crea una nueva figura y devuelve una tupla que contiene la figura (fig) y un conjunto de ejes (ax) que se utilizarán para crear y personalizar gráficos en la figura.



**La figura (fig)** es el contenedor principal donde se colocarán los gráficos. Puedes imaginarlo como el lienzo en el que se dibujarán los gráficos.


**El conjunto de ejes (ax)** representa los ejes coordenados en los que se trazarán los datos. Los ejes actúan como contenedores para los diferentes elementos del gráfico, como líneas, barras, puntos, etiquetas, etc. Cada gráfico individual se crea dentro de un conjunto de ejes.

**Graficos de dispersion: scatter**


La función scatter() en Matplotlib se utiliza para crear gráficos de dispersión (scatter plots). Un gráfico de dispersión muestra puntos individuales en un plano cartesiano, donde cada punto representa una combinación de valores para dos variables diferentes,permite detectar patrones o tendencias.


Podemos dibujar los puntos (lista_x[i],lista_y[i]) sin unir por líneas, apareciendo los puntos de forma dispersa.

````python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 1, 6, 3])
#Son dos arrays NumPy que representan las coordenadas 'x' e 'y' de los puntos que se dibujarán en el gráfico de dispersión.
tamanos = np.array([30, 50, 80, 20, 100])
colores = ['red', 'green', 'blue', 'yellow', 'purple']
#marker='x' , '*' , entre otros
plt.scatter(x, y, s=tamanos, c=colores, marker='*')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
#Estas funciones agregan etiquetas a los ejes 'x' e 'y', respectivamente.
plt.title('Gráfico de dispersión')
plt.grid(True)
plt.show()
````

````python
# Generar datos de ejemplo dentro del rango específico
# Generar 50 números aleatorios en el rango [0, 4)
x = np.random.uniform(0, 4, 50)  # Valores x aleatorios entre 0 y 4 con distribucion uniforme
y = np.random.uniform(0, 4, 50)  # Valores y aleatorios entre 0 y 4 con distribucion uniforme


# Plotear los puntos
plt.scatter(x, y,marker='*')
plt.title("Ploteo de puntos en el rango de 0 al 4")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.xlim(0, 4)  # Limitar el rango del eje x
plt.ylim(0, 4)  # Limitar el rango del eje y
plt.grid(True)
plt.show()
````

````python
# Generar datos aleatorios para horas de estudio y puntaje en un examen
np.random.seed(42)  # Para reproducibilidad
horas_estudio = np.random.rand(50) * 10  # 50 valores aleatorios entre 0 y 10 horas
puntaje_examen = 50 + 4 * horas_estudio + np.random.normal(0, 5, 50)  # puntaje = 50 + 4 * horas_estudio + ruido gaussiano

# Crear el diagrama de dispersión

plt.scatter(horas_estudio, puntaje_examen, label='Datos')

# Agregar etiquetas de ejes y título
plt.xlabel('Horas de estudio')
plt.ylabel('Puntaje en el examen')
plt.title('Diagrama de dispersión: Horas de estudio vs Puntaje en el examen')
plt.grid(True)
plt.show()
````

**Dibujando varias gráficas superpuestas**

````python

# Datos de ventas diarias
ventas_diarias = np.array([
    #lu,mar,mier,jue,vie,saba,domi
    [20, 15, 25, 30, 18, 22, 30],  # Producto A
    [12, 10, 14, 8, 15, 5, 1],   # Producto B
    [35, 28, 32, 30, 26, 24, 30],  # Producto C
    [40, 38, 45, 42, 39, 41, 37]   # Producto D
])

# Días de la semana
dias_semana = ['Lun', 'Mar', 'Mie', 'Jue', 'Vie', 'Sab', 'Dom']

# Gráfico de líneas para mostrar las ventas diarias de cada producto
fig, ax = plt.subplots()


for i, producto in enumerate(['A', 'B', 'C', 'D']):
    #i=0,produc=lista[indice] -> A
    #i=1,produc=B
    #..
    ax.plot(dias_semana, ventas_diarias[i], label='Producto ' + producto)

ax.set_xlabel('Días de la semana')
ax.set_ylabel('Ventas diarias')
ax.set_title('Ventas diarias de productos')
ax.legend()
plt.show()
````

**Graficar Funciones**

**Funcion Lineal**

````python
# Crear valores de x en el rango de -5 a 5
x = np.linspace(-5, 5, 100) #crear un array x que contiene 100 puntos igualmente espaciados en el intervalo entre
# Calcular los valores de y utilizando la función y = 3x - 2
y = 3 * x - 2
#Suma +
#Resta -
# *
# **

# Crear una figura y ejes para el gráfico
fig, ax = plt.subplots()

# Trazar la función y = 3x - 2
ax.plot(x, y, label='y = 3x - 2')

# Agregar etiquetas de ejes y título
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Gráfico de la función y = 3x - 2')

# Mostrar la leyenda
ax.legend()

# Mostrar el gráfico
plt.grid(True)
plt.show()
````

**Funcion cuadratica**

````python
# Valores de x en el rango de -5 a 5
x = np.linspace(-5, 5, 100)

# Calcula los valores de y utilizando la función cuadrática y = x^2 - 4
y_cuadratica = x**2 - 4

# Graficar la función cuadrática

plt.plot(x, y_cuadratica, label='y = x^2 - 4')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de la función cuadrática y = x^2 - 4')
plt.legend()
plt.grid(True)
plt.show()
````

**Función normal (gaussiana):**

````python
import matplotlib.pyplot as plt
import numpy as np

# Valores de x en el rango de -5 a 5
x = np.linspace(-5, 5, 100)

# Media y desviación estándar de la función normal
media = 0
desviacion_estandar = 1

# Calcula los valores de y utilizando la función normal (gaussiana)
y_normal = (1 / (desviacion_estandar * np.sqrt(2 * np.pi))) * np.exp(-(x - media)**2 / (2 * desviacion_estandar**2))

# Graficar la función normal (gaussiana)
plt.plot(x, y_normal, label='Función normal (gaussiana)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de la función normal (gaussiana)')
plt.legend()
plt.grid(True)
plt.show()
````

**Gráfico de torta**

Un gráfico de torta es una representación visual que muestra la proporción de diferentes partes o categorías en un conjunto de datos. Cada categoría se representa como una porción del círculo, y el tamaño de cada porción corresponde a su proporción con respecto al total. Es una forma rápida y fácil de visualizar cómo se distribuyen las partes individuales dentro de un conjunto más grande.


Nota :
Al utilizar autopct='%1.1f%%', se mostrarán los porcentajes correspondientes a cada categoría con un formato de número decimal con una precisión de 1 dígito después del punto decimal, seguido del símbolo de porcentaje '%'.

````python
#Grafica de torta
categorias = ['A', 'B', 'C', 'D']
valores = [10, 25, 15, 30]

plt.pie(valores, labels=categorias, autopct='%1.1f%%')
plt.title('Gráfico de torta')
plt.show()
````

````python
# Contar la cantidad de muestras por variedad
variety_counts = iris['variety'].value_counts()

# Crear el gráfico de torta
plt.pie(variety_counts, labels=variety_counts.index, autopct='%1.2f%%', startangle=140)
plt.title('Distribución de Muestras por Variedad en el Conjunto de Datos Iris')

# Mostrar el gráfico
plt.axis('equal')  # Para que el gráfico de torta sea circular
plt.show()
````

````python


# Generar datos aleatorios para horas de estudio y puntaje en un examen
np.random.seed(42)  # Para reproducibilidad
horas_estudio = np.random.rand(50) * 10  # 50 valores aleatorios entre 0 y 10 horas
puntaje_examen = 50 + 4 * horas_estudio + np.random.normal(0, 5, 50)  # puntaje = 50 + 4 * horas_estudio + ruido gaussiano

# Crear el diagrama de dispersión
plt.scatter(horas_estudio, puntaje_examen, label='Datos')

# Agregar etiquetas de ejes y título
plt.xlabel('Horas de estudio')
plt.ylabel('Puntaje en el examen')
plt.title('Diagrama de dispersión: Horas de estudio vs Puntaje en el examen')




# Ajustar un modelo lineal
coef = np.polyfit(horas_estudio, puntaje_examen, 1)  # Ajuste de un polinomio de grado 1 (modelo lineal)
puntaje_predicho = coef[1] + coef[0] * horas_estudio  # Puntaje predicho utilizando el modelo

# Agregar la línea de regresión al gráfico
plt.plot(horas_estudio, puntaje_predicho, color='red', label='Modelo lineal')

# Mostrar el diagrama de dispersión y la línea de regresión
plt.legend()
plt.grid(True)
plt.show()

# Estimar el puntaje en función del número de horas de estudio
horas_estudio_desconocidas = np.array([5, 8, 10])
puntaje_estimado = coef[1] + coef[0] * horas_estudio_desconocidas
print("Puntaje estimado para las siguientes horas de estudio:")
for horas, puntaje in zip(horas_estudio_desconocidas, puntaje_estimado):
    print(f"Horas de estudio: {horas}, Puntaje estimado: {puntaje:.2f}")
````
