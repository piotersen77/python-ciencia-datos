**Matrices en Numpy**





````python
import numpy as np
````

**Creación de matrices**


```
np.array(lista): Crea un arreglo multidimensional desde una lista o tupla.
np.zeros(shape, dtype): Crea una matriz de ceros con dimensiones especificadas.
np.ones(shape, dtype): Crea una matriz de unos con dimensiones especificadas.
np.diag(lista_diagonal): Crea una matriz diagonal con los elementos de lista_diagonal
np.identity(N,dtype): Crea una matriz identidad (unos en la diagonal y ceros en otros lugares).
```



````python
#matriz
matriz=np.array([[1,3,4],[5,1,6],[5,6,7]])

print(matriz)

#matriz con zeros
matriz_zeros = np.zeros((3,3))
print(matriz_zeros)


#Matriz ones
matriz_one = np.ones((3,3))
print(matriz_one)

#matriz diagonal

matriz_dia = np.diag([1,3,4,5])
print(matriz_dia)

#matriz identidad
matriz_id = np.identity(3)
print(matriz_id)








````

**Indexacion**

````python
               #    0       1      2
matriz=np.array([[1,3,4],[5,1,6],[5,6,7]])

print('primer ele',matriz[1,1])
print('primera fila',matriz[0,:])#[fila,columna]
print('primera fila',matriz[:,0])

#Indexacion booleana
matriz = np.array([[0,3 ,8],[5, 6 ,6],[8, 9 ,1]])

print('matriz orgional',matriz)
# Definir una condición (por ejemplo, queremos los elementos mayores que 5)
condicion = matriz > 5
# Aplicar la condición a la matriz y obtener un nuevo array con los elementos que cumplen la condición
nueva_matriz = matriz[condicion]

print('nueva matriz',nueva_matriz)
````

**Operaciones con matrices**

**Suma de matrices**


La suma de dos matrices A y B (ambas con las mismas dimensiones) es que cada elemento de la matriz resultante C tendrá el valor de la suma de los elementos correspondientes de A y B.

````python
matriz_a=np.array([[1,3],[5,1]])
matriz_b=np.array([[5,3],[5,1]])
print('La matriz resultante es ',matriz_a + matriz_b)
````

**Resta de matrices**

La definición de la resta de dos matrices A y B (ambas con las mismas dimensiones) es que cada elemento de la matriz resultante C tendrá el valor de la resta de los elementos correspondientes de A y B

````python
matriz_a=np.array([[1,3],[5,1]])
matriz_b=np.array([[5,3],[5,1]])

print('La matriz resultante es ',matriz_a - matriz_b)
````

**Suma,resta y multiplicacion por un escalar**

````python
matriz_a=np.array([[1,3,4],[5,1,6],[5,6,7]])
result=matriz_a - 2
print('La matriz resultante es ',result)
````

**Matriz traspuesta**

 Es una operación que intercambia filas por columnas en una matriz, resultando en una nueva matriz con sus filas y columnas invertidas



```
np.transpose(matriz)
```



````python
matriz=np.array([[1,3,4],[5,1,6]])
print('Matriz A',matriz.shape)
matriz_t = np.transpose(matriz)
print('Matriz A^T',matriz_t.shape)

````

**Matriz inversa**

La matriz inversa es una operación importante en el álgebra lineal que nos permite "desinvertir" una matriz, de manera que al multiplicarla por su inversa, obtenemos la matriz de identidad.

````python
A = np.array([[1, 2],
              [3, 4]])


A_inv= np.linalg.inv(A)

print(A_inv)


````

**Producto de matrices**

Número de columnas de A igual al número de filas de B: Para realizar el producto de matrices A * B, el número de columnas de la matriz A debe ser igual al número de filas de la matriz B.

Para los que no se acuerdan la multiplicacion de matrices :

[Videito](https://www.youtube.com/watch?v=Tjrm3HsqBXE)

````python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Producto de matrices utilizando np.dot() o el operador @
nueva_matriz = np.dot(A,B)

print(nueva_matriz)
````

**Concatenacion**

La concatenación de matrices es una operación que combina dos o más matrices para formar una nueva matriz más grande.

````python
# Definir las matrices A y B
A = np.array([[1, 2]
              ,[3, 4]])

B = np.array([[5, 6],
              [7, 8]])

#[1,2,5,6]
#[3,4,7,8]

# Concatenación horizontal (columnas)
concatenacion_horizontal = np.concatenate((A, B), axis=1)



# Concatenación vertical (filas)
concatenacion_vertical = np.concatenate((A, B), axis=0)
#[1,2]
#[3,4]
#[5,6]
#[7,8]

print("Concatenación horizontal:")
print(concatenacion_horizontal)
print("Concatenación vertical:")
print(concatenacion_vertical)
````

**Funcion sum**


```
np.sum(matriz, axis=None, dtype=None, keepdims=False, initial=0, where=True)

```
Matriz: El array o matriz sobre el cual se va a realizar la suma.


axis (opcional): El eje a lo largo del cual se realizará la suma. Si no se especifica, se sumarán todos los elementos del array y se devolverá un único valor.


where (opcional): Una condición booleana o máscara que indica qué elementos se deben incluir en la suma. Solo se sumarán los elementos donde where es True. Por defecto, se incluyen todos los elementos.


````python

#Suma total de todos los elementos
matriz=np.array([
                [1,3,4], #1+3+4 = 8
                [5,1,6]])
                #6,4,10

#Suma todos los elementos de la matriz
sumaEle=np.sum(matriz)
print('Suma total matriz',sumaEle)

#Suma por fila
sumaxFila= np.sum(matriz,axis=1)
print('La suma por fila',sumaxFila)

#suma por columnas
sumaxColumna= np.sum(matriz,axis=0)

print('La suma por columna',sumaxColumna)

````

**Ejercicio**

Dada la siguiente dataset representado por una matriz, donde cada fila representa la cantidad que se vendio en referencia a un solo producto durante toda la semana; mientras que cada columna representa la venta total en un dia de la semana.Obtener la siguiente información detallada:

````python
# Datos de ventas diarias de 4 productos durante una semana

ventas_diarias = np.array(
    [
    #Lun,Mar,Mie,Jue,Vie,Sab,Dom
    [20, 15, 25, 30, 18, 22, 24],   #Producto A
    [12, 10, 14, 8, 15, 18, 16],    #Producto B
    [35, 28, 32, 30, 26, 24, 30],   #Producto C
    [40, 38, 45, 42, 39, 41, 37]    #Producto D
    ]
)
#Venta por dia
ventas_por_dia = np.sum(ventas_diarias,axis=0)
print(f'Esta son la ventas de cada dia de la semana{ventas_por_dia}')



#Vetas por producto

ventas_por_dia = np.sum(ventas_diarias,axis=1)
print(f'Esta son la ventas de cada producto en la semana {ventas_por_dia}')
````
